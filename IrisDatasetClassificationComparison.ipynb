```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from urllib.request import urlopen , Request
```

```python
d=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
```

```python
d.head()
```

```python
from sklearn.utils import shuffle
shuffle(d)
```

```python
from sklearn.model_selection import train_test_split

X = d.iloc[:, :-1]

Y = d.iloc[:, -1:]

from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
Y=lb.fit_transform(Y)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=0)
```

```python
print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='gini',max_depth=2)
dtc.fit(X_train,Y_train)
```

```python
y_pred=dtc.predict(X_test)
print(X_test)
print(Y_test)
print(y_pred)
plt.plot(y_pred,Y_test)
plt.xlabel('predictedvalue')
plt.ylabel('testvalue')
plt.show()
```

```python
score=dtc.score(X_test,Y_test)
print(score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=8,weights='uniform')
kn.fit(X_train,Y_train)
import warnings
warnings.filterwarnings('ignore')
```

```python
score=kn.score(X_test,Y_test)
print(score)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
score=lr.score(X_test,Y_test)
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=9,criterion='entropy',max_depth=None)
rf.fit(X_train,Y_train)
```

```python
score=rf.score(X_test,Y_test)
print(score)
```
